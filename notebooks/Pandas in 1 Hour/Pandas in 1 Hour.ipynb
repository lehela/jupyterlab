> #### Data Profiling
> 
>  Profile the number of unique values in a given column

```python
sales.nunique()
```

> Small numbers indicate that these columns could be referenced as "categories" to save memory

```python
sales["Platform"] = sales["Platform"].astype("category")
sales["Genre"] = sales["Genre"].astype("category")
```

> Note that the memory usage has decreased now by about 30%

```python
sales.info()
```

```python
sales["Genre"].value_counts() 
sales["Genre"].value_counts(normalize = True) * 100
```

> ## Melting / Unpivoting

```python
sales = sales.reset_index()
sales.head()
```

> Note how the numeric columns essentially are all amounts

> These columns can be unpivoted to 2 columns of "Region" and "Sales"

>  Region    Sales
    ----    ------
>   NA        0.15

>   EU        0.10


```python
sales_unpivot = sales.melt(
    id_vars=["Name", "Publisher"],
    value_vars=["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"],
    var_name="Region",
    value_name="Sales"
)
sales_unpivot.head()
```

```python
sales_unpivot[sales_unpivot["Name"] == "'98 Koshien"]
```
